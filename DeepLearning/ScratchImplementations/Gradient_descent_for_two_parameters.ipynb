```python
X = [0.5,2.5]
Y = [0.2,0.9]
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(w,b,x):
    return 1.0/(1.0 + np.exp(-(w*x + b)))
```

```python
def error(w,b):
    err = 0
    for x,y in zip(X,Y):
        y_pred = f(w,b,x)
        err += 0.5 * (y_pred - y)**2
    return err
```

```python
def grad_b(w,b,x,y):
    fx = f(w,b,x)
    return (fx-y) * fx * (1-fx)
```

```python
def grad_w(w,b,x,y):
    fx = f(w,b,x)
    return (fx-y) * fx * (1-fx) * (x)
```

```python
def plot3d(x,y):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    p,q = np.meshgrid(x,y)
    z = error(p,q)
    ax.plot_surface(p, q, z,cmap = 'viridis')
    # ax.plot_surface(x, y, z)
    ax.set_xlabel('w')
    ax.set_ylabel('b')
    ax.set_zlabel('Loss')

    ax.set_xlim(-10,10)
    ax.set_ylim(-10,10)
    ax.set_zlim(0,0.5)
    plt.show()
    return
```

```python
def gradient_descent(epochs = 1000):
    w,b,eta = 0.0,-8.0,1.0
    w_ls = [w]
    b_ls = [b]
    for i in range(epochs):
        dw,db = 0.0,0.0
        n = len(X)
        for x,y in zip(X,Y):
            dw += grad_w(w,b,x,y) * (1/n)
            db += grad_b(w,b,x,y) * (1/n)
        w = w - eta * dw
        b = b - eta * db
        w_ls.append(w)
        b_ls.append(w)
    print(w,b,error(w,b))
    plot3d(np.asarray(w_ls),np.asarray(b_ls))
```

```python
gradient_descent()
```
