# Class

```python
class person:
    """This is a basic class structure"""
    def __init__(self,name,age = 20,weight = 45):
        self.name = 'keep smiling ' + name
        self.age = age
        self.weight = weight
        print('Stupid')

    def is_heavy(self):
        return self.weight > 50
```

```python
class Male(person):
    def __init__(self,name,gender = 'male',weight = 45):
        super().__init__(name)
        self.gender = gender
        print(self.name)
    def printage(self):
        print(f'Your age is {self.age}')
    def is_heavy(self):
        return self.weight > 40
```

# Debugging


```python
def factorial(x):
    if (not isinstance(x,int)):
        print('Give valid integer input')
        return -1
        
    if x == 1 or x ==0:
        return 1
    else:
        return x * factorial(x-1)
```

```python
factorial(6)
```

```python
def plot_multiple_vecs(vecs):
    c = ['r','b','g','y','skyblue']
    i = 0
    for vec in vecs:
        plt.quiver(vec[0],vec[1],vec[2],vec[3],scale_units = 'xy',angles = 'xy',scale = 1,color = c[i%5])
        i = i+1
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    plt.show()
```

```python
vecs = [np.array([0,0,3,4]),np.array([0,0,-3,4]),np.array([0,0,-5,-4]),np.array([0,0,3,-2])]
```

```python
plot_multiple_vecs(vecs)
```

```python
vecs[0] + vecs[1]
```

```python
plot_multiple_vecs([vecs[1],vecs[2],vecs[1] + vecs[2]])
```

# vector dot product

```python
a = np.asarray([5,6])
b = np.asarray([-2,7])
```

$\vec{a}\cdot\vec{b} = |\vec{a}||\vec{b}|\cos(\theta) = a_xb_x + a_yb_y$

```python
a_dot_b = np.dot(a,b)
print(a_dot_b)
```

```python
np.linalg.norm(b)
```

```python
a_b = np.dot(a,b)/np.linalg.norm(b)
print(a_b)             # projection of a on b
```

```python
a_b_vec = (a_b/np.linalg.norm(b))*b
print(a_b_vec)
```

```python
plot_multiple_vecs([[0,0,a[0],a[1]],[0,0,b[0],b[1]],[0,0,a_b_vec[0],a_b_vec[1]]])
```
