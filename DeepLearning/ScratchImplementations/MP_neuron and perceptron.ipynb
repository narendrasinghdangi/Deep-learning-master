## MP_Neuron_Class

```python
class MPNeuron:

    def __init__(self):
        self.b = None

    def model(self,X):
        return sum(X) >= self.b

    def predict(self,X):
        y_pred = []
        for x in X:
            y_pred.append(self.model(x))
        return np.array(y_pred)
    
    def fit(self,X,Y):
        accuracy = {}

        for b in range(X.shape[1]+1):
            self.b = b
            Y_pred = self.predict(X)
            accuracy[b] = accuracy_score(Y_pred, Y)
        
        optimal_b = max(accuracy,key = accuracy.get)
        self.b = optimal_b
        print('Optimal value of b is', optimal_b)
        print('Highest accuracy is', accuracy[optimal_b])
    
```

```python
mp_neuron = MPNeuron()
mp_neuron.fit(x_binarized_train,y_train)
```

```python
y_test_pred = mp_neuron.predict(x_binarized_test)
accuracy = accuracy_score(y_test_pred,y_test)
print(accuracy)
```

## Perceptron class with epochs

```python
class Perceptron:

    def __init__(self):
        self.w = None
        self.b = None
    
    def model(self,x):
        return 1 if (np.dot(self.w, x) >= self.b) else 0
    
    def predict(self,x):
        y = []
        for i in x:
            y.append(self.model(i))
        return np.array(y)
    
    def fit(self,X,Y,epochs = 1):

        self.w = np.random.randn(X.shape[1])
        self.b = np.random.randn(1)
        accuracy = {}

        for i in range(epochs):
            for x,y in zip(X,Y):
                y_pred = self.model(x)
                if y == 1 and y_pred == 0:
                    self.w = self.w + x
                    self.b = self.b + 1
                if y == 0 and y_pred == 1:
                    self.w = self.w - x
                    self.b = self.b - 1
            
            accuracy[i] = accuracy_score(self.predict(X),Y)

        print(max(accuracy.values()))
        plt.plot(np.array(list(accuracy.values())).astype(float))
        plt.show()     
```

## Hyper Parameter Tuning

### checkpoints

```python
class Perceptron:

    def __init__(self):
        self.w = None
        self.b = None
    
    def model(self,x):
        return 1 if (np.dot(self.w, x) >= self.b) else 0
    
    def predict(self,x):
        y = []
        for i in x:
            y.append(self.model(i))
        return np.array(y)
    
    def fit(self,X,Y,epochs = 1):

        self.w = np.ones(X.shape[1])
        self.b = 0
        accuracy = {}
        max_accuracy = 0

        for i in range(epochs):
            for x,y in zip(X,Y):
                y_pred = self.model(x)
                if y == 1 and y_pred == 0:
                    self.w = self.w + x
                    self.b = self.b + 1
                if y == 0 and y_pred == 1:
                    self.w = self.w - x
                    self.b = self.b - 1

            accuracy[i] = accuracy_score(self.predict(X),Y)
            if accuracy[i] > max_accuracy:
                max_accuracy = accuracy[i]
                chkptw = self.w
                chkptb = self.b

        self.w = chkptw
        self.b = chkptb 

        print(max_accuracy)
        plt.plot(np.array(list(accuracy.values())).astype(float))
        plt.show()     
```

### Learning rate

```python
class Perceptron:

    def __init__(self):
        self.w = None
        self.b = None
    
    def model(self,x):
        return 1 if (np.dot(self.w, x) >= self.b) else 0
    
    def predict(self,x):
        y = []
        for i in x:
           result = self.model(i)
           y.append(result)
        return np.array(y)
    
    def fit(self,X,Y,epochs = 1, lr = 1):

        self.w = np.ones(X.shape[1])
        self.b = 0

        accuracy = {}
        max_accuracy = 0

        for i in range(epochs):
            for x,y in zip(X,Y):
                y_pred = self.model(x)

                if y == 1 and y_pred == 0:
                    self.w = self.w + lr * x
                    self.b = self.b + lr * 1
                elif y == 0 and y_pred == 1:
                    self.w = self.w - lr * x
                    self.b = self.b - lr * 1
                    
            accuracy[i] = accuracy_score(self.predict(X),Y)

            if accuracy[i] > max_accuracy:
                max_accuracy = accuracy[i]
                chkptw = self.w
                chkptb = self.b

        self.w = chkptw
        self.b = chkptb
        print(max_accuracy)
        plt.plot(np.array(list(accuracy.values())).astype(float))
        plt.show()     
```

### Animating how weight's change in the algorithm

```python
class Perceptron:

    def __init__(self):
        self.w = None
        self.b = None
    
    def model(self,x):
        return 1 if (np.dot(self.w, x) >= self.b) else 0
    
    def predict(self,x):
        y = []
        for i in x:
           result = self.model(i)
           y.append(result)
        return np.array(y)
    
    def fit(self,X,Y,epochs = 1, lr = 1):

        self.w = np.ones(X.shape[1])
        self.b = 0

        accuracy = {}
        max_accuracy = 0
        wt_matrix = []

        for i in range(epochs):
            for x,y in zip(X,Y):
                y_pred = self.model(x)

                if y == 1 and y_pred == 0:
                    self.w = self.w + lr * x
                    self.b = self.b + lr * 1
                elif y == 0 and y_pred == 1:
                    self.w = self.w - lr * x
                    self.b = self.b - lr * 1

            wt_matrix.append(self.w)

            accuracy[i] = accuracy_score(self.predict(X),Y)
            if accuracy[i] > max_accuracy:
                max_accuracy = accuracy[i]
                chkptw = self.w
                chkptb = self.b

        self.w = chkptw
        self.b = chkptb
        print(max_accuracy)
        plt.plot(np.array(list(accuracy.values())).astype(float))
        plt.show()     

        return np.array(wt_matrix)
```

```python
perceptron = Perceptron()
```

```python
wt_matrix = perceptron.fit(x_train,y_train,100,1)
```
